Every year the Nobel Prize is given to scientists and scholars in the categories chemistry, literature, physics, physiology or medicine, economics, and peace.

<h3>Number of Prizes Awarded Over Time</h3>

```python
prize_per_year = df_data.groupby(by='year').count().prize
prize_per_year
```

```python
moving_average = prize_per_year.rolling(window=5).mean()
moving_average
```

```python
np.arange(1900, 2021, step=5)
```

```python
plt.figure(figsize=(16,8), dpi=200)
plt.title('Number of Nobel Prizes Awarded per Year', fontsize=18)
plt.yticks(fontsize=14)
plt.xticks(ticks=np.arange(1900, 2021, step=5),
           fontsize=14,
           rotation=45)

ax = plt.gca() # get current axis
ax.set_xlim(1900, 2020)

ax.scatter(x=prize_per_year.index,
           y=prize_per_year.values,
           c='dodgerblue',
           alpha=0.7,
           s=100,)

ax.plot(prize_per_year.index,
        moving_average.values,
        c='crimson',
        linewidth=3,)

plt.show()
```

<b>Conclusion:</b>
<p>It is noticeable that the number of prizes has increased over the years.</p>

<h3>Are More Prizes Shared Than Before?</h3>

```python
yearly_avg_share = df_data.groupby(by='year').agg({'share_pct': pd.Series.mean})
share_moving_average = yearly_avg_share.rolling(window=5).mean()
share_moving_average
```

```python
plt.figure(figsize=(16,8), dpi=200)
plt.title('Number of Nobel Prizes Awarded per Year', fontsize=18)
plt.yticks(fontsize=14)
plt.xticks(ticks=np.arange(1900, 2021, step=5),
           fontsize=14,
           rotation=45)

ax1 = plt.gca()
ax2 = ax1.twinx() # create second y-axis
ax1.set_xlim(1900, 2020)

ax2.invert_yaxis()

ax1.scatter(x=prize_per_year.index,
           y=prize_per_year.values,
           c='dodgerblue',
           alpha=0.7,
           s=100,)

ax1.plot(prize_per_year.index,
        moving_average.values,
        c='crimson',
        linewidth=3,)

# Adding prize share plot on second axis
ax2.plot(prize_per_year.index,
        share_moving_average.values,
        c='grey',
        linewidth=3,)

plt.show()
```

<b>Conclusion:</b>
<p>It can be seen on the graph that increased number of shared prizes over the years.</p>

<h3>Age at Time of Award throughout History</h3>

```python
plt.figure(figsize=(8,4), dpi=200)
with sns.axes_style("whitegrid"):
    sns.regplot(data=df_data,
                x='year',
                y='winning_age',
                lowess=True,
                scatter_kws = {'alpha': 0.4},
                line_kws={'color': 'black'}).set(title="Age at Time of Award throughout History")

plt.show()
```

<p><b>Conclusion:</b></p>
<p>A weak positive correlation tells us that the age trending up over time..</p> 

```python
plt.figure(figsize=(8,4), dpi=200)
sns.boxplot(data=df_data, x="winning_age", y="category").set(title="Winning Age Across the Nobel Prize Categories")
plt.show
```

<p><b>Conclusion:</b><p>
<ul>
  <li>The physics category has the longest "whiskers", which means the biggest difference in winning age.</li>
  <li>The Economics category has the shortest difference in winning age.</li>
  <li>The oldest average winners are in the Literature and Economics categories.</li>
  <li>The youngest average winners are in the Physics category.</li>

<h3>What are the winning age trends in each category?</h3>

```python
plt.figure(figsize=(8,4), dpi=200)
sns.lmplot(
    data=df_data, x="year", y="winning_age",
    col="category", height=6,
    facet_kws=dict(sharex=False, sharey=False), lowess=True, hue='category'
)
plt.show()
```

<p><b>Conclusion:</b></p>
<ul>
  <li>Chemistry, Literature, Medicine and Physics category has the age trending up.</li>
  <li>The peace category has the age trending down.</li>
</ul>



```python
plt.figure(figsize=(8,4), dpi=200)
with sns.axes_style("whitegrid"):
    sns.lmplot(data=df_data,
               x='year',
               y='winning_age',
               hue='category',
               lowess=True,
               aspect=2,
               scatter_kws={'alpha': 0.5},
               line_kws={'linewidth': 5})

plt.show()
```

<b>Conclusion:</b>
<ul>
    <li><b>Physics, Chemistry and Medicine:</b> These fields have shown a noticeable increase in the average age of Nobel Prize winners over time. In physics particularly, there has been a significant shift from an average age below 50 to over 70. This could be attributed to several factors such as the increasing complexity of research, longer education and career paths and advancements in technology requiring more time for significant contributions.</li>
    <p></p>
    <li><b>Economics:</b> Economics, being a relatively newer category compared to physics, chemistry and medicine has exhibited a more stable trend in terms of the age of its laureates. This could be because the field of economics has matured more recently and its laureates may not have experienced the same generational shifts or changes in research dynamics as the older disciplines.</li>
    <p></p>
    <li><b>Peace Prize:</b> The Peace Prize shows a contrasting trend where winners are getting younger over time. This could be due to a greater emphasis on youth-led movements, activism and initiatives in promoting peace and human rights. Younger individuals may also be more actively involved in social and political movements, leading to their recognition for Nobel Peace Prizes.</li>
</ul>

The discrepancy between the trends observed in scatter plots and box plots underscores the importance of analyzing data from different perspectives. While scatter plots with best-fit lines can provide insights into the changing trends over time, box plots offer a snapshot of the entire dataset's distribution, highlighting variations and outliers that might not be apparent in trendlines alone. It's essential to consider both types of visualizations to fully understand the nuances within the data.
